```python
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import brier_score_loss

#Tuples: ()
n_models = [(df_Eurosc_II_xgboost_test, 'Xgboost') ]
#verified that calibration does not affect the AUC score values for Random Forest beta calibration 

# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
brier = dict()
for i in range(len(n_models)):
    #extra [0] is needed index the tuple to obtain dataframe 
    fpr[i], tpr[i], _ = roc_curve(n_models[i][0]['mtly'], n_models[i][0].iloc[:,0])  #get first column of dataframe i.e. meanProb
    roc_auc[i] = auc(fpr[i], tpr[i])
    #above two lines is equivalent to roc_auc_score
    brier[i] = brier_score_loss(n_models[i][0]['mtly'], n_models[i][0].iloc[:,0])
```

```python
#based on: https://stackoverflow.com/questions/19124239/scikit-learn-roc-curve-with-confidence-intervals
from sklearn.metrics import roc_auc_score

#print("Original ROC area: {:0.3f}".format(roc_auc_score(y_true, y_pred)))

n_bootstraps = 1000
rng_seed = 42  # control reproducibility
bootstrapped_scores = [[] for i in range(len(n_models))] # list in list: https://stackoverflow.com/questions/12791501/python-initializing-a-list-of-lists

for i in range(len(n_models)):
    rng = np.random.RandomState(rng_seed)
    for j in range(n_bootstraps):
        # bootstrap by sampling with replacement on the prediction indices
        indices = rng.randint(0, len(n_models[i][0].iloc[:,0]), len(n_models[i][0].iloc[:,0])) #(param1 lowest values; param2: largest - 1 value to sample; param3: no. samples)
        if len(np.unique(n_models[i][0]['mtly'][indices])) < 2:
            # We need at least one positive and one negative sample for ROC AUC
            # to be defined: reject the sample
            continue

        score = roc_auc_score(n_models[i][0]['mtly'][indices], n_models[i][0].iloc[:,0][indices])
        bootstrapped_scores[i].append(score)
        print("Bootstrap #{} ROC area: {:0.4f}".format(j + 1, score))
```

```python
!pip install scipy
from scipy.stats import sem, t
from scipy import mean, stats

from itertools import cycle
import matplotlib.ticker as plticker

# Plot all ROC curves
plt.figure()
fig=plt.figure(figsize=(12,8), dpi= 300, facecolor='w', edgecolor='k')
ax = fig.add_subplot(1, 1, 1)

lw = 2

colors = cycle(["blue","green","brown","red","salmon","#7CD3CF","purple"])
for i, color in zip(range(len(n_models)), colors):
    
    #calculate 95% CI
    sorted_scores = np.array(bootstrapped_scores[i])
    sorted_scores.sort()

    # Computing the lower and upper bound of the 95% confidence interval 
    # You can change the bounds percentiles to 0.025 and 0.975 to get
    # a 95% confidence interval instead.
    # http://www.stat.ucla.edu/~rgould/110as02/bsci
    
    #https://www.kite.com/python/answers/how-to-compute-the-confidence-interval-of-a-sample-statistic-in-python
    confidence_level = 0.95
    degrees_freedom = len(sorted_scores) - 1
    sample_mean = np.mean(sorted_scores)
    sample_standard_error = stats.sem(sorted_scores)

    confidence_interval = t.interval(confidence_level, degrees_freedom, sample_mean, sample_standard_error)
    confidence_upper = confidence_interval[1] #upper
    confidence_lower = confidence_interval[0] #lower
    
    
    plt.plot(fpr[i], tpr[i], color=color, lw=lw,
             label='{0} AUC {1:0.4f} \n (95CI {2:0.4f} - {3:0.4f})'.format(n_models[i][1], roc_auc[i], confidence_lower, confidence_upper))

plt.plot([0, 1], [0, 1], 'k--', lw=lw)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('SBFS feature model: Holdout Validation ROC curve')
plt.legend(loc="lower right")

#Spacing between each line

minor_ticks = np.arange(0, 1, 0.1)
ax.set_xticks(minor_ticks, minor=True)
ax.set_yticks(minor_ticks, minor=True)

ax.grid(which='minor', alpha=0.2)

ax.grid(which='major')
plt.show()
```
